```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\adina\Downloads\Sales_add.csv')
```

```python
data.head()
```

```python
from scipy .stats import ttest_ind
```

```python
#1
sales_before = data['Sales_before_digital_add(in $)']
sales_after = data['Sales_After_digital_add(in $)']
```

```python
t_stat, p_val = ttest_ind(sales_before, sales_after)
```

```python
p_val
```

```python
if p_val < 0.05:
    print("Null Hypothesis Rejected")
else:
    print("Null Hypothesis Accepted")
```

```python
plt.subplots(figsize=(10, 6))
plt.plot(data['Month'], sales_before, label='Sales Before')
plt.plot(data['Month'], sales_after, label='Sales After')
plt.xlabel('Month')
plt.ylabel('Sales (in $)')
plt.title('Sales Before and After Digital Marketing')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#### So there is increase in sales after stepping into digital marketing
